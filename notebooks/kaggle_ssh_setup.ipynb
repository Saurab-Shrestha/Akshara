```python
# Install openssh-server and set root password
!apt-get install -y -q openssh-server
PASSWORD = 'change-me'  # set your own password before running
!echo f'root:{PASSWORD}' | chpasswd

# Allow root login
!sed -i 's/#PermitRootLogin.*/PermitRootLogin yes/' /etc/ssh/sshd_config
!sed -i 's/#PasswordAuthentication.*/PasswordAuthentication yes/' /etc/ssh/sshd_config
!service ssh start
print('SSH server running')
```

```python
import subprocess, time, json

# Start ngrok tunnel in background
subprocess.Popen(['./ngrok', 'tcp', '22'])
time.sleep(3)

# Get the tunnel address
import urllib.request
tunnels = json.loads(urllib.request.urlopen('http://localhost:4040/api/tunnels').read())
addr = tunnels['tunnels'][0]['public_url'].replace('tcp://', '')
host, port = addr.split(':')

print(f'\nConnect with:')
print(f'  ssh root@{host} -p {port}')
print(f'  password: <the PASSWORD you set in cell-6>')
print(f'\nOr add to ~/.ssh/config:')
print(f'  Host kaggle')
print(f'    HostName {host}')
print(f'    Port {port}')
print(f'    User root')
```
